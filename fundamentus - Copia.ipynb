```python
#from tabulate import tabulate
#import re

import pandas as pd
import bs4 as bs
import string
import requests 
import warnings
warnings.filterwarnings('ignore')
```

```python
print("BeautifulSoup ->", bs.__version__)
print("pandas ->", pd.__version__)

url = 'https://www.fundamentus.com.br/resultado.php'
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
response = requests.get(url,headers=headers)
#soup = bs.BeautifulSoup(response.content, 'html.parser')
#table = soup.find_all('table')[0]
#df = pd.read_html(str(table), decimal=',', thousands='.')[0]

df = pd.read_html(response.text, decimal=',', thousands='.')[0]
df.head()
```

```python
#df.columns =[column.replace('/', "") for column in df.columns]
#df.columns =[column.replace(". ", "") for column in df.columns]
#df.columns =[column.replace(".", "") for column  in df.columns]
#df.columns =[column.replace(" ", "") for column  in df.columns]
```

```python
#df.ROE = df.ROE.apply(lambda x: x.replace('%',' ').replace(',','.'))
for coluna in ['Div.Yield', 'Mrg Ebit','Mrg. Líq.', 'ROIC', 'ROE', 'Cresc. Rec.5a']:
    df[coluna] = df[coluna].str.replace('.', '').replace(',', '.').rstrip('%').astype('float') * 100
```

```python
df.ROE= pd.to_numeric(df.ROE, errors='coerce')
df.ROIC =pd.to_numeric(df.ROIC, errors='coerce')
df['DivYield'] =pd.to_numeric(df['DivYield'], errors='coerce')
df['MrgEbit'] =pd.to_numeric(df['MrgEbit'], errors='coerce')
df['MrgLíq'] =pd.to_numeric(df['MrgLíq'], errors='coerce')

```
