```python
#from tabulate import tabulate
#import re

import pandas as pd
import bs4 as bs
import string
import requests 
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'http://www.fundamentus.com.br/resultado.php'
headers = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

response = requests.get(url,headers=headers)
df = pd.read_html(response.text, decimal=',', thousands='.')[0]
```

```python
for coluna in ['Div.Yield', 'Mrg Ebit','Mrg. Líq.', 'ROIC', 'ROE', 'Cresc. Rec.5a']:
    df[coluna] = df[coluna].str.replace('.', '').str.replace(',', '.').str.rstrip('%').astype('float') * 100
```

```python
df.shape
```

```python
df = df[df['Liq.2meses'] > 1000000]
```

```python
df.shape
```

```python
rank = pd.DataFrame()
rank['pos'] = range(1,151)
rank['EV/EBIT'] = df[df['EV/EBIT'] > 0].sort_values(by=['EV/EBIT'])['Papel'][:150].values
rank['ROIC'] = df.sort_values(by=['ROIC'],ascending=False)['Papel'][:150].values
```

```python
rank
```

```python
a = rank.pivot_table(columns='EV/EBIT', values='pos')
b = rank.pivot_table(columns='ROIC', values='pos')
```

```python
t = pd.concat([a,b])
```

```python
ranking = t.dropna(axis=1).sum()
```
